# Installing Python program

## Setup - window users

 1. Visit https://www.python.org/
 


## Click on downloads and select Windows


## A list of python releases for 32 bit/64 bit will be available, select the appropriate release based on your system processing unit


## Once downloaded, run the setup file on your machine


## A pop up window will arise, select the check box "Add python __(i.e. 3.9) to PATH" and click on "Install Now"


## Click on "Yes" when asked "Do you want to allow this app to make changes to your device"?

## The installation was successful


# Setup - MAC users

## Click on downloads and select MacOS


## A list of python releases for 32 bit/64 bit will be available, select the appropriate release based on your system processing unit
